### Analysis and Visualization

```python
cancelled_df = df['is_canceled'].value_counts(normalize=True)
cancelled_df
```

```python
chart_data = df['is_canceled'].value_counts(normalize=True).round(2)
plt.figure(figsize=(3,4))
plt.title("reservation cancellation chart")
plt.bar(['Not canceled', 'canceled'], chart_data, width=0.5)

for i in range(len(chart_data)):
    plt.text(i, chart_data[i]/2, chart_data[i], ha='center')

plt.show()
```

```python
resort_hotel_df = df[df['hotel']=='Resort Hotel']
resort_hotel_df1 = resort_hotel_df['is_canceled'].value_counts()
resort_hotel_df1
```

```python
resort_hotel_cancelation_percent_df = resort_hotel_df['is_canceled'].value_counts(normalize=True).round(2)

plt.figure(figsize=(3,4))
plt.title('resort hotel cancellation chart')
plt.bar(['Not canceled', 'canceled'], resort_hotel_cancelation_percent_df, width=0.5)

for i in range(len(resort_hotel_cancelation_percent_df)):
    plt.text(i, resort_hotel_cancelation_percent_df[i]/2, resort_hotel_cancelation_percent_df[i], ha='center')
    
plt.show()
```

```python
city_hotel_df = df[df['hotel']=='City Hotel']
city_hotel_df1 = city_hotel_df['is_canceled'].value_counts()
city_hotel_df1

city_hotel_cancelation_percent_df = city_hotel_df['is_canceled'].value_counts(normalize=True).round(2)

plt.figure(figsize=(3,4))
plt.title('city hotel cancellation chart')
plt.bar(['Not canceled', 'canceled'], city_hotel_cancelation_percent_df, width=0.5)

for i in range(len(city_hotel_cancelation_percent_df)):
    plt.text(i, city_hotel_cancelation_percent_df[i]/2, city_hotel_cancelation_percent_df[i], ha='center')
    
plt.show()
```

check price is making impact on cancellation

```python
resort_hotel_df
```

```python
start_date = pd.to_datetime('2015-08-31', format="%Y-%m-%d")
resort_hotel_price_df = resort_hotel_df[resort_hotel_df['reservation_status_date']>start_date].groupby('reservation_status_date')[['adr']].mean()
city_hotel_price_df = city_hotel_df[city_hotel_df['reservation_status_date']>start_date].groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15,7))
plt.title('Average daily rate in resort and city hotel', fontsize=20)
plt.plot(resort_hotel_price_df.index, resort_hotel_price_df['adr'], label='resort_hotel')
plt.plot(city_hotel_price_df.index, city_hotel_price_df['adr'], label='city_hotel')
plt.legend(fontsize=8)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(15,5))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df)
plt.title('Reservation status month wise', size=15)
plt.xlabel('month')
plt.ylabel('cancelation_count')
plt.legend(['not_canceled','canceled'])
plt.show()
```

check daily rate month wise to check impact on cancellation

```python
plt.figure(figsize=(15,5))
plt.title('ADR per month', size=15)
bar_data = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month',y='adr', data=bar_data)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('ADR per month', size=15)
bar_data = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month',y='adr', data=bar_data)

for i in range(len(bar_data)):
    plt.text(i, bar_data['adr'][i], bar_data['adr'][i], ha='center')
    
plt.show()
```

From above graph we can conculde that average monthly booking rate is highest in the january month and lowest in the september month. For the same month in the previous chart we have seen highest cancellation in january month and lowest in the september month.
Booking rate directly impacts cancellation rate, if prices are higher then cancellation chances are also higher.

#### Top 10 Country wise cancellation 

```python
canceled_data = df[df['is_canceled']==1]
country_wise_canceled_df = canceled_data['country'].value_counts()[0:10]
country_wise_canceled_df
```

```python
plt.figure(figsize=(5,6))
plt.title('Top 10 Country\'s cancelation rate' )
plt.pie(country_wise_canceled_df,labels=country_wise_canceled_df.index, autopct='%.2f')
plt.show()
```

PRT, GBR and ESP are top 3 countries where cancellation is highest

```python
end_date = pd.to_datetime('2017-07-01', format="%Y-%m-%d")
adr_canceled_data = canceled_data[(canceled_data['reservation_status_date']>start_date) & (canceled_data['reservation_status_date']<end_date)].groupby('reservation_status_date')[['adr']].sum()

not_canceled_data = df[df['is_canceled']==0]
adr_not_canceled_data = not_canceled_data[(not_canceled_data['reservation_status_date']>start_date) & (not_canceled_data['reservation_status_date']<end_date)].groupby('reservation_status_date')[['adr']].sum()

plt.figure(figsize=(12,4.5))
plt.title('Average daily rate chart', size=8)
plt.plot(adr_canceled_data.index, adr_canceled_data['adr'], label='cancelled')
plt.plot(adr_not_canceled_data.index, adr_not_canceled_data['adr'], label='not cancelled')
plt.legend()
plt.show()
```
